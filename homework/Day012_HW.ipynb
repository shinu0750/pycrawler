## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
news = soup.find(class_="part_list_2").find_all("h3")

for new in news:
    print(new.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
Class = soup.find(class_="part_list_2").find_all("em")
ClassList = []

for item in Class:
    ClassList.append(item.text)
    
ClassSet = set(ClassList)
ClassList = list(ClassSet)

ClassNAList = []
for i in range(len(ClassList)):
    ClassNAList.append([])

for item in Class:
    i = ClassList.index(item.text)
    ClassNAList[i].append(item.parent.text.replace("\u3000",""))
    
ClassDict = {}
for i in range(len(ClassNAList)):
    ClassDict[ClassList[i]] = ClassNAList[i]

ClassDict
```

### ③ 哪一個類別的文章最多

```python
Count = []
for i in range(len(ClassDict.keys())):
    Count.append(len(ClassDict[list(ClassDict.keys())[i]]))

print("Max Catelog: " + str(ClassList[Count.index(max(Count))]))
```
